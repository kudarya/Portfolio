- Frequency (–ß–∞—Å—Ç–æ—Ç–∞) ‚Äî –ö–∞–∫ —á–∞—Å—Ç–æ –æ–Ω–∏ —Å–æ–≤–µ—Ä—à–∞—é—Ç –ø–æ–∫—É–ø–∫–∏?
- Monetary Value (–î–µ–Ω–µ–∂–Ω–∞—è —Ü–µ–Ω–Ω–æ—Å—Ç—å) ‚Äî –°–∫–æ–ª—å–∫–æ –æ–Ω–∏ —Ç—Ä–∞—Ç—è—Ç?

–ó–∞—Ç–µ–º —Ä–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å —Å–µ–≥–º–µ–Ω—Ç—ã –≤ —Ä–∞–∑—Ä–µ–∑–µ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏ –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–π —Ç–æ–≤–∞—Ä–æ–≤.

### RFM-–∞–Ω–∞–ª–∏–∑

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ –∫ —Å–µ–≥–º–µ–Ω—Ç–∞—Ü–∏–∏.

```python
# –æ—Ç–¥–µ–ª—å–Ω–æ –≤—ã–≤–µ–¥–µ–º –¥–∞—Ç—É –∑–∞–∫–∞–∑–∞,–∏–Ω–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –∏ —Å—É–º–º—É –∑–∞–∫–∞–∑–∞ 
df_rfm = df_clean\
                .groupby(['date','customer_id'], as_index=False)\
                .agg({'sum': 'sum'})

df_rfm
```

```python
# –ù–∞–π–¥–µ–º —á–∞—Å—Ç–æ—Ç—É –∑–∞–∫–∞–∑–æ–≤ (Frequency)
frequency = df_rfm['customer_id'].value_counts().reset_index()
frequency.rename(columns={'index': 'customer_id', 'customer_id': 'frequency'}, inplace=True)
frequency
```

```python
# –ò–∑—É—á–∏–º –æ–¥–Ω–æ—Ä–æ–¥–Ω–æ—Å—Ç—å –¥–∞–Ω–Ω—ã—Ö
frequency.describe(percentiles = [.20, .40, .60, .80, .90, .99])
```

99% –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –¥–µ–ª–∞–ª–∏ 2 –∏–ª–∏ –º–µ–Ω–µ–µ –∑–∞–∫–∞–∑–∞, 90%  1 –∑–∞–∫–∞–∑, —Ç–æ –µ—Å—Ç—å –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –≤ –º–∞–≥–∞–∑–∏–Ω–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏ —Å–æ–≤–µ—Ä—à–∞—é—Ç 1 –∑–∞–∫–∞–∑ –∏ –Ω–µ –≤–æ–∑–≤—Ä–∞—â–∞—é—Ç—Å—è.

```python
# –ù–∞–π–¥–µ–º –æ–±—â—É—é —Å—É–º–º—É –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–∫—É–ø–∞—Ç–µ–ª—é
total_sum_by_customer = df_rfm.groupby(by='customer_id').sum().groupby(level=[0]).cumsum()\
                                            .sort_values(by='sum', ascending=False)
total_sum_by_customer
```

```python
# –ù–∞–π–¥–µ–º –¥–∞—Ç—ã –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–∫—É–ø–∞—Ç–µ–ª—é, —á—Ç–æ–±—ã –æ–ø–∏—Å–∞—Ç—å –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å Recency
order_by_last_date = df_rfm.sort_values('date').groupby('customer_id').last()
order_by_last_date.rename(columns={'date': 'order_last_date'}, inplace=True)
order_by_last_date = order_by_last_date.drop(['sum'], axis=1) # —É–¥–∞–ª–∏–º –Ω–µ–Ω—É–∂–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü sum
order_by_last_date
```

```python
# –û–±—ä–µ–¥–∏–Ω–∏–º –¥–≤–µ —Ç–∞–±–ª–∏—Ü—ã —Å –æ–±—â–µ–π —Å—É–º–º–æ–π –∑–∞–∫–∞–∑–∞ –∏ –¥–∞—Ç–æ–π –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –∑–∞–∫–∞–∑–∞ –ø–æ –∏–Ω–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä—É 
# –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
rfm = order_by_last_date.merge(total_sum_by_customer, on='customer_id')
rfm
```

```python
# –ü–æ—Å–º–æ—Ç—Ä–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
rfm.describe(percentiles = [.20, .40, .60, .80, .90, .95, .97])
```

97% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å–¥–µ–ª–∞–ª–∏ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ —Å—É–º–º—É 5462.4 –∏ –º–µ–Ω—å—à–µ, –ø–æ—ç—Ç–æ–º—É –æ–±—ä–µ–¥–∏–Ω–∏–º –≤—Å–µ—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –ø–æ—Ç—Ä–∞—Ç–∏–≤—à–∏—Ö –±–æ–ª–µ–µ 5000 –¥.–µ. –≤ —Å–µ–≥–º–µ–Ω—Ç most_monetary. –° —ç—Ç–∏–º —Å–µ–≥–º–µ–Ω—Ç–æ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ—Ä–∞–±–æ—Ç–∞—Ç—å –æ—Ç–¥–µ–ª—å–Ω–æ —Å —Ü–µ–ª—å—é —É–¥–µ—Ä–∂–µ–Ω–∏—è –∏ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–∏—è –ø—Ä–µ—Ñ–µ—Ä–µ–Ω—Ü–∏–π –∫–∞–∫ –≤–∏–ø-–∫–ª–∏–µ–Ω—Ç–∞–º. 

```python
most_monetary = rfm.query('sum >= 5000')
print(most_monetary.shape)
most_monetary = most_monetary.sort_values(by='order_last_date')
most_monetary
```

–°–µ–≥–º–µ–Ω—Ç most_monetary —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ 91 –ø–æ–∫—É–ø–∞—Ç–µ–ª—è, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–ª–∏ –µ–≥–æ –ø–æ –ø–æ—Å–ª–µ–¥–Ω–µ–π –¥–∞—Ç–µ –∑–∞–∫–∞–∑–∞, —á—Ç–æ–±—ã –≤—ã–¥–µ–ª–∏—Ç—å –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ —Ç—Ä–µ–±—É—é—Ç –¥–µ–π—Å—Ç–≤–∏–π –¥–ª—è –≤–æ–∑–≤—Ä–∞—Ç–Ω–æ—Å—Ç–∏. –ü—Ä–∏–º–µ–º, —á—Ç–æ –Ω–∞ –º–æ–º–µ–Ω—Ç –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –¥–∞—Ç–∞ - 31.10.2019 (–¥–µ–Ω—å –ø–æ—Å–ª–µ–¥–Ω–µ–π –∑–∞–ø–∏—Å–∏ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ) –∏ –∏—Å—Å–ª–µ–¥—É–µ–º, –∫–∞–∫ –¥–æ–ª–≥–æ –Ω–µ —Å–æ–≤–µ—Ä—à–∞–ª–∏ –∑–∞–∫–∞–∑ –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏ –∏–∑ —Å–µ–≥–º–µ–Ω—Ç–∞ most_monetary.

```python
now = '2019-10-31-23'
rfm['current_date'] = now
rfm['current_date'] = pd.to_datetime(rfm['current_date'], format='%Y-%m-%d-%H')
rfm['churn_time'] = rfm['current_date'] - rfm['order_last_date']
rfm
```

```python
# –ü–æ—Å–º–æ—Ç—Ä–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
rfm.describe()
```

–ü–æ–ª–æ–≤–∏–Ω–∞ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π —Å–¥–µ–ª–∞–ª–∞ –∑–∞–∫–∞–∑ –≤ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ –º–µ–∂—É 212 –¥–Ω—è–º–∏ –∏ 395 –¥–Ω–µ–π, —Ç–æ –µ—Å—Ç—å –æ—Ç 7 –º–µ—Å—è—Ü–µ–≤ –¥–æ –≥–æ–¥–∞, –æ—Ç—Ç–æ–∫ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã–π, –Ω–æ –µ—Å–ª–∏ –≤—Å–ø–æ–º–Ω–∏—Ç—å, —á—Ç–æ –º–∞–≥–∞–∑–∏–Ω –ø–æ–¥–≤–µ—Ä–∂–µ–Ω —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏, —Ç–æ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –ø–æ–Ω—è—Ç–Ω–æ, —á—Ç–æ –∫–∞–∫ —Ä–∞–∑ –æ—Ç –≤—Ä–µ–º–µ–Ω–∏ –Ω–∞—á–∞–ª–∞ –¥–∞—á–Ω–æ–≥–æ —Å–µ–∑–æ–Ω–∞ (–∞–ø—Ä–µ–ª—å-–º–∞–π) –∏ –¥–æ –Ω–∞—á–∞–ª–∞ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –∫ –Ω–æ–≤–æ–≥–æ–¥–Ω–∏–º –ø—Ä–∞–∑–¥–Ω–∏–∫–∞–º –∫–∞–∫ —Ä–∞–∑ –ø—Ä–æ—Ö–æ–¥–∏—Ç –æ–∫–æ–ª–æ 7 –º–µ—Å—è—Ü–µ–≤. –ù–æ —Ç–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –≤–µ—Ä–Ω—É–ª–∏—Å—å –≤ —Ç–µ—á–µ–Ω–∏–µ —ç—Ç–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏, —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ –æ—Å—Ç–∞–ª–∏—Å—å —á–µ–º-—Ç–æ –Ω–µ–¥–æ–≤–æ–ª—å–Ω—ã –∏–ª–∏ –∑–∞–±—ã–ª–∏ –ø—Ä–æ –º–∞–≥–∞–∑–∏–Ω. –ü–æ—Å–º–æ—Ç—Ä–∏–º, —Å–∫–æ–ª—å–∫–æ —Ç–∞–∫–∏—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π.

```python
# –î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É rfm —Å–æ–ª–±–µ—Ü —Å —á–∞—Å—Ç–æ—Ç–æ–π –∑–∞–∫–∞–∑–æ–≤ –∏ –≤—ã–¥–µ–ª–∏–º —Ç–µ—Ö, –∫—Ç–æ –Ω–µ —Å–æ–≤–µ—Ä—à–∞–ª –∑–∞–∫–∞–∑ 211 –¥–Ω–µ–π
rfm_final = rfm.merge(frequency, on='customer_id').sort_values(by='sum', ascending=False)
churn_customers = rfm_final.query('churn_time > "211 days"')
churn_customers
```

–¢–∞–∫–∏—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –æ—á–µ–Ω—å –º–Ω–æ–≥–æ. –≠—Ç–æ –±—É–¥–µ—Ç –æ—Ç–¥–µ–ª—å–Ω—ã–π —Å–µ–≥–º–µ–Ω—Ç churn_customers - –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ –Ω–∞–¥–æ –ø–æ—Ä–∞–±–æ—Ç–∞—Ç—å —Å —Ü–µ–ª—å—é –≤–µ—Ä–Ω—É—Ç—å –∏—Ö –≤ –º–∞–≥–∞–∑–∏–Ω.

–£ –Ω–∞—Å –ø–æ–ª—É—á–∏–ª–∞—Å—å —Ç–∞–±–ª–∏—Ü–∞ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π —Å –ø–æ—Å–ª–µ–¥–Ω–µ–π –¥–∞—Ç–æ–π –∑–∞–∫–∞–∑–∞, —á–∞—Å—Ç–æ—Ç–æ–π –∑–∞–∫–∞–∑–æ–≤ –∏ –æ–±—â–µ–π —Å—É–º–º–æ–π –∑–∞–∫–∞–∑–æ–≤, –∞ —Ç–∞–∫–∂–µ —Å –¥–Ω—è–º–∏ –ø—Ä–µ–±—ã–≤–∞–Ω–∏—è –≤ –æ—Ç—Ç–æ–∫–µ. –ü—Ä–∏—Å—Ç—É–ø–∏–º –∫ RFM-–∞–Ω–∞–ª–∏–∑—É. –î–æ–±–∞–≤–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∑–Ω–∞—á–µ–Ω–∏–µ Monetary, —É—á–∏—Ç—ã–≤–∞—è —Ç–æ, —á—Ç–æ –≤—Å–µ–≥–æ –ª–∏—à—å 3% –æ—Ç –≤—Å–µ—Ö –∫–æ–Ω—Ç—Ä–∞–≥–µ–Ω—Ç–æ–≤ —Å–¥–µ–ª–∞–ª–∏ –∑–∞–∫–∞–∑–æ–≤ –≤—Å–µ–≥–æ –Ω–∞ —Å—É–º–º—É –±–æ–ª–µ–µ 5281.68 –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–∏–º –ø–æ —Å–µ–≥–º–µ–Ω—Ç–∞–º. –ü—Ä–∏—Å–≤–æ–∏–º –∑–Ω–∞–µ–Ω–∏–µ –ú = 1, –µ—Å–ª–∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª—å –ø–æ—Ç—Ä–∞—Ç–∏–ª –±–æ–ª–µ–µ 5281.68, –ú = 2, –µ—Å–ª–∏ –ø–æ—Ç—Ä–∞—Ç–∏–ª –º–µ–Ω–µ–µ 5281.68  –∏ –ú = 3, –µ—Å–ª–∏ –ø–æ—Ç—Ä–∞—Ç–∏–ª –º–µ–Ω–µ–µ 704 (–º–µ–¥–∏–∞–Ω–∞ –ø–æ –æ–±—â–µ–π —Å—É–º–º–µ –∑–∞–∫–∞–∑–æ–≤)

```python
rfm_final['M'] = [1 if x > 5281.68 else 0 for x in rfm_final['sum']]
rfm_final.loc[rfm_final['sum'] <= 5281.68, 'M'] = 2
rfm_final.loc[rfm_final['sum'] <= 704, 'M'] = 3
rfm_final
```

–ü—Ä–∏—Å–≤–æ–∏–º –∏–Ω–¥–µ–∫—Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ F: F = 1, –µ—Å–ª–∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª—å —Å–¥–µ–ª–∞–ª –±–æ–ª–µ–µ 2 –∑–∞–∫–∞–∑–æ–≤, F = 2, –µ—Å–ª–∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª—å —Å–¥–µ–ª–∞–ª –º–µ–Ω–µ–µ 2 –∑–∞–∫–∞–∑–æ–≤.

```python
rfm_final['F'] = [1 if x >= 2 else 0 for x in rfm_final['frequency']]
rfm_final.loc[rfm_final['frequency'] < 2, 'F'] = 2
rfm_final
```

–î–æ–±–∞–≤–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–∞–≤–Ω–æ—Å—Ç–∏ –∑–∞–∫–∞–∑–∞ Recency –Ω–∞ –æ—Å–Ω–æ–≤–µ –∏–∑—É—á–µ–Ω–∏—è churn_time, –ø–æ–º–Ω—è –æ –ø–æ–¥–≤–µ—Ä–∂–µ–Ω–Ω–æ—Å—Ç–∏ –º–∞–≥–∞–∑–∏–Ω–∞ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏ –∏ –≤—ã—á–∏—Å–µ–Ω–Ω–æ–º—É –≤—Ä–µ–º–µ–Ω–∏, –∑–∞ –∫–æ—Ç–æ—Ä–æ–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª—å –≥–∏–ø–æ—Ç–µ—Ç–∏—á–µ—Å–∫–∏ –¥–æ–ª–∂–µ–Ω –≤–µ—Ä–Ω—É—Ç—å—Å—è - 7 –º–µ—Å—è—Ü–µ–≤. –ü—Ä–∏—Å–≤–æ–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ R = 1, –µ—Å–ª–∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª—å –≤–µ—Ä–Ω—É–ª—Å—è –≤ —Ç–µ—á–µ–Ω–∏–µ 7 –º–µ—Å—è—Ü–µ–≤, R = 2, –µ—Å–ª–∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª—å –Ω–µ –≤–µ—Ä–Ω—É–ª—Å—è –≤ —Ç–µ—á–µ–Ω–∏–∏ 7 –º–µ—Å—è—Ü–µ–≤.

```python
delta = timedelta(days=211, hours=0, minutes=0)
rfm_final['R'] = [1 if x < delta else 0 for x in rfm_final['churn_time']]
rfm_final.loc[rfm_final['churn_time'] >= delta, 'R'] = 2
rfm_final
```

–ü–æ–ª—É—á–∏–ª–∏ –∏—Ç–æ–≥–æ–≤—É—é —Ä–∞–∑–±–∏–≤–∫—É –ø–æ RFM c —É—á–µ—Ç–æ–º –Ω–µ–æ–¥–Ω–æ—Ä–æ–¥–Ω–æ—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö –≤ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ. –ú–æ–∂–Ω–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–µ —Å–µ–≥–º–µ–Ω—Ç—ã:


1. –¢–µ, –∫—Ç–æ —á–∞—Å—Ç–æ –ø–æ–∫—É–ø–∞–ª–∏ –∏ –ø—Ä–∏–Ω–µ—Å–ª–∏ –Ω–∞–±–æ–ª—å—à—É—é –≤—ã—Ä—É—á–∫—É (M = 1, F = 1)
2. –¢–µ, –∫—Ç–æ –ø—Ä–∏–Ω–µ—Å–ª–∏ –±–æ–ª—å—à—É—é –≤—ã—Ä—É—á–∫—É, –Ω–æ –ø–æ–∫—É–ø–∞–ª–∏ –Ω–µ —á–∞—Å—Ç–æ (M = 1, F = 2)
3. –¢–µ, –∫—Ç–æ —á–∞—Å—Ç–æ –ø–æ–∫—É–ø–∞–ª–∏ –∏ –ø—Ä–∏–Ω–µ—Å–ª–∏ –º–∞–ª—É—é –≤—ã—Ä—É—á–∫—É (M = 3, F = 1)
4. –¢–µ, –∫—Ç–æ –≤–æ–∑–≤–∞—Ä–∞—â–∞–ª–∏—Å—å –≤ —Ç–µ—á–µ–Ω–∏–µ 7 –º–µ—Å—è—Ü–µ–≤ –∏ –ø—Ä–∏–Ω–µ—Å–ª–∏ –Ω–∞–∏–±–æ–ª—å—à—É—é –≤—ã—Ä—É—á–∫—É (–ú = 1, R = 1)

- H_1: –í—ã—Ä—É—á–∫–∞ (sum) —Å –ø—Ä–æ–¥–∞–∂ —Ç–æ–≤–∞—Ä–æ–≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –•–æ–∑—è–π—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ç–æ–≤–∞—Ä—ã ‚â† –≤—ã—Ä—É—á–∫–∞ (sum) —Å –ø—Ä–æ–¥–∞–∂ —Ç–æ–≤–∞—Ä–æ–≤ –¥—Ä—É–≥–∏—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π


- alpha = 0.05

```python
a = df_clean.query('product_category == "–•–æ–∑—è–π—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ç–æ–≤–∞—Ä—ã"')
b = df_clean.query('product_category != "–•–æ–∑—è–π—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ç–æ–≤–∞—Ä—ã"')

#value_1 = df_clean.loc[(df_clean['product_category'] == '–•–æ–∑—è–π—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ç–æ–≤–∞—Ä—ã'), 'sum']
#value_2 = df_clean.loc[(df_clean['product_category'] != '–•–æ–∑—è–π—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ç–æ–≤–∞—Ä—ã'), 'sum']

value_1 = a.groupby('product')['sum'].agg('sum')
value_2 = b.groupby('product')['sum'].agg('sum')

results = st.ttest_ind(
    value_1, 
    value_2, equal_var = False) # –≤—ã–∑–æ–≤ –º–µ—Ç–æ–¥–∞ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑—ã

alpha = .05 #–∑–∞–¥–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

print(results.pvalue) # –≤—ã–≤–æ–¥ –∑–Ω–∞—á–µ–Ω–∏—è p-value –Ω–∞ —ç–∫—Ä–∞–Ω 
# —É—Å–ª–æ–≤–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä —Å –≤—ã–≤–æ–¥–æ–º —Å—Ç—Ä–æ–∫–∏ —Å –æ—Ç–≤–µ—Ç–æ–º
if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```

–¢–µ, –∫—Ç–æ —á–∞—Å—Ç–æ –ø–æ–∫—É–ø–∞–ª–∏ –∏ –ø—Ä–∏–Ω–µ—Å–ª–∏ –º–∞–ª—É—é –≤—ã—Ä—É—á–∫—É (M = 3, F = 1)
